## 0. INSPEÇÃO INICIAL (LIMPEZA E TRATAMENTO)

```python
# IMPORTANDO BIBLIOTECA PANDAS PARA ANÁLISE DE DADOS
import pandas as pd
import re
```

```python
# IMPORTANDO BASE DE DADOS DE ARQUIVO CSV
df = pd.read_csv('train.csv')
df.head(5)
```

```python
# TROCANDO NOMES DAS COLUNAS
colunas = {'ID': 'ID',
                     'Delivery_person_ID': 'ID_Entregador',
                     'Delivery_person_Age': 'Idade_Entregador',
                     'Delivery_person_Ratings': 'Avaliacoes_Entregador',
                     'Restaurant_latitude': 'Latitude_Restaurante',
                     'Restaurant_longitude': 'Longitude_Restaurante',
                     'Delivery_location_latitude': 'Latitude_Entrega',
                     'Delivery_location_longitude': 'Longitude_Entrega',
                     'Order_Date': 'Data_Pedido',
                     'Time_Orderd': 'Hora_Pedido',
                     'Time_Order_picked': 'Hora_Entrega',
                     'Weatherconditions': 'Condicoes_Climaticas',
                     'Road_traffic_density': 'Densidade_Trafego',
                     'Vehicle_condition': 'Estado_Veiculo',
                     'Type_of_order': 'Tipo_Pedido',
                     'Type_of_vehicle': 'Tipo_Veiculo',
                     'multiple_deliveries': 'Entregas_Multiplos',
                     'Festival': 'Festival',
                     'City': 'Cidade',
                     'Time_taken(min)': 'Tempo_Levado(min)'}
df.rename(columns=colunas, inplace=True)
df.head(5)
```

```python
# CRIANDO CÓPIA DO DATAFRAME PRINCIPAL PARA ALTERAÇÕES
df1 = df.copy()
```

```python
# DESCRIÇÃO GERAL DO DATAFRAME
df1.describe()
```

```python
# AVALIAÇÃO TÉCNICA DOS DADOS DO DATAFRAME
df1.info()
```

```python
# REALIZANDO CONVERSÃO NECESSÁRIA DE TIPOS PARA IDADE, AVALIAÇÕES E MULTIPLAS ENTREGAS
# Removendo registros com NaN 
linhas_validas = (df1['Idade_Entregador'] != 'NaN ') & (df1['Avaliacoes_Entregador'] != 'NaN ') & (df1['Entregas_Multiplos'] != 'NaN ')
df1 = df1.loc[linhas_validas, :].copy()

# Conversão com .astype()
df1['Idade_Entregador'] = df1['Idade_Entregador'].astype(int)
df1['Avaliacoes_Entregador'] = df1['Avaliacoes_Entregador'].astype(float)
df1['Entregas_Multiplos'] = df1['Entregas_Multiplos'].astype(int)

# CONVERTENDO COLUNA DE DATA
df1['Data_Pedido'] = pd.to_datetime(df1['Data_Pedido'], format='%d-%m-%Y')

# RESETANDO ÍNDICES PARA RECONFIGURAÇÃO DO DF
df1 = df1.reset_index(drop=True)

df1.dtypes
```

```python
# LIMPEZA DO CAMPO DE TEMPO:
df1['Tempo_Levado(min)'] = df1['Tempo_Levado(min)'].str.extract('(\d+)').astype(int)
```

```python
# LIMPEZA DE ESPAÇOS EM BRANCO JUNTO AOS DADOS DOS REGISTROS USANDO .strip()
# ID (0), ID_Entregador(1), Densidade_Trafego(12), Tipo_Pedido(14) 
# Tipo_Veiculo (15), Festival(17), Cidade(18) 

# Índices das colunas que você deseja aplicar strip
indices_colunas = [0, 1, 12, 14, 15, 17, 18]

# Aplicar strip a cada elemento nas colunas selecionadas
for coluna in indices_colunas:
    df1.iloc[:, coluna] = df1.iloc[:, coluna].astype(str).str.strip()
```

# 1. Qual a média de idade dos entregadores por cidade?

```python
# Agrupando cidade
df_cidades = df1.groupby('Cidade')

# Calculando média de idade dos entregadores
df_1 = df_cidades['Idade_Entregador'].mean().reset_index(name='Media de Idade dos Entregadores')
df_1
```

# 2. Qual a mediana de idade dos entregadores por cidade?

```python
# Reutilizando df_grupamento com cidades agrupadas
# Calculando mediana de idade dos entregadores
df_2 = df_cidades['Idade_Entregador'].median().reset_index(name='Mediana de Idade dos Entregadores')
df_2
```

# 3. Qual o número total de entregas feitas por densidade de tráfego?

```python
# Agrupando densidades de tráfego
df_densidades = df1.groupby('Densidade_Trafego')

# Calculando total de entregas por cada densidade
df_3 = df_densidades['ID'].count().reset_index(name='Total de Entregas')

# Mostrando resultados do maior pra menor
df_3.sort_values(by=['Total de Entregas'], ascending=False).reset_index(drop=True)
```

# 4. Qual a média das avaliações das entregas feitas por densidade de tráfego??

```python
# Reaproveitando agrupamento anterior de densidades de trânsito
# Calculando média das avaliações de entrega
df_4 = df_densidades['Avaliacoes_Entregador'].mean().reset_index(name='Media de Avaliacoes')

# Exibindo em ordem decrescente
df_4.sort_values(by=['Media de Avaliacoes'], ascending=False).reset_index(drop=True)
```

# 5. Qual a quantidade de entregas feitas por cidade?

```python
# Reutilizando agrupamento de cidades
# Calculando quantidade de entregas
df_5 = df_cidades['ID'].count().reset_index(name='Quantidade de Entregas')

# Exibindo em ordem decrescente
df_5.sort_values(by=['Quantidade de Entregas'], ascending=False).reset_index(drop=True)
```

# 6. Qual a média, mediana e desvio padrão do tempo das entregas feitas por cidade e por tipo de veículo?

```python
# Agrupando cidades e tipos de veículos
df_cv = df1.groupby(['Cidade', 'Tipo_Veiculo'])

# Utilizando função de agregação agg() para métricas desejadas
df_6 = df_cv['Tempo_Levado(min)'].agg(['mean', 'median', 'std']).reset_index()

# Mapeando nomes
nomes = {
    'mean': 'Media de Tempo de Entrega (min)',
    'median': 'Mediana de Tempo de Entrega (min)',
    'std': 'Desvio Padrao de Tempo de Entrega (min)'
}

# Renomeando colunas
df_6 = df_6.rename(columns=nomes)

# Ordenando pela média
df_6.sort_values(by=['Media de Tempo de Entrega (min)'], ascending=False).reset_index(drop=True)
```

# 7. Qual a primeira data de entrega por cada cidade?

```python
# Reaproveitando agrupamento anterior de cidades
# Calculando data mais antiga
df_7 = df_cidades['Data_Pedido'].min().reset_index()
df_7
```

# 8. Qual a última data de entrega por cada cidade??

```python
# Reaproveitando agrupamento anterior de cidades
# Calculando data mais antiga
df_8 = df_cidades['Data_Pedido'].max().reset_index()
df_8
```

# 9. Qual a média de avaliações das entregas feitas por condições climáticas?

```python
# Agrupando condições climáticas
df_cc = df1.groupby('Condicoes_Climaticas')

# Calculando média de avaliações das entregas
df_9 = df_cc['Avaliacoes_Entregador'].mean().reset_index(name='Media de Avaliacoes de Entregas')

# Exibindo em ordem decrescente
df_9.sort_values(by=['Media de Avaliacoes de Entregas'], ascending=False).reset_index(drop=True)
```

# 10. Qual o valor da avaliação mais baixa feita por tipo de condição climática e por densidade de tráfego?

```python
# Agrupando condicoes climaticas e densidades de trafego
df_cc_dt = df1.groupby(['Condicoes_Climaticas', 'Densidade_Trafego'])

# Calculando avaliações
avaliacoes_cc_dt = df_cc_dt['Avaliacoes_Entregador'].min().reset_index()
avaliacoes_cc_dt
```

# 11. Qual a média, mediana e desvio padrão das avaliações por cada tipo de condições de veículos?

```python
# Agrupando condições de veículos
df_cv = df1.groupby('Estado_Veiculo')

# Utilizando função de agregação agg() para calcular métricas solicitadas
df_11 = df_cv['Avaliacoes_Entregador'].agg(['mean', 'median', 'std']).reset_index()

df_11
```

# 12. Quantas entreagas foram feitas no Festival?

```python
# Contagem da qtd de entregas na condição de festival
condicao = df1['Festival'] == 'Yes'
qtd_entregas_festival = len(df1.loc[condicao, :])

print(f"Quantidade de entregas no festival: {qtd_entregas_festival}")
```

# 13. Qual a avaliação média das entregas feitas no Festival?

```python
# Condicao de festival
condicao = df1['Festival'] == 'Yes'

# Registros com festival
df_festival = df1.loc[condicao, :].copy()

# Calculando média
print("Média de avaliações de entregas feitas em festival: ")
print(df_festival['Avaliacoes_Entregador'].mean())
```

# 14. Qual a menor avaliação feita em uma entrega no Festival e por cidade?

```python
# Reaproveitando df com registros em festival
# Agrupando cidade e festival
df_cf = df_festival.groupby(['Cidade', 'Festival'])

# Calculando menor avaliação por cidade em entregas no festival
df_14 = df_cf['Avaliacoes_Entregador'].min().reset_index(name='Menor Avaliacao')

# Exibindo em ordem crescente
df_14.sort_values(by=['Menor Avaliacao'], ascending=True).reset_index(drop=True)
```

# 15. Qual a maior avaliação feita por tipo de pedido?

```python
# Agrupando tipos de pedidos
df_tp = df1.groupby('Tipo_Pedido')

# Calculando maior avaliação
df_15 = df_tp['Avaliacoes_Entregador'].max().reset_index(name='Maior Avaliação')

# Exibindo em ordem decrescente
df_15.sort_values(by=['Maior Avaliação'], ascending=False).reset_index(drop=True)
```
